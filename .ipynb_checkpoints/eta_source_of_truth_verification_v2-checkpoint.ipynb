```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
from datetime import datetime
import hashlib, platform, json

plt.style.use('seaborn-v0_8')
pd.options.display.max_rows = 50

CSV_PATH = 'datasets/cleaned_eta_logs.csv'

# Lock to dataset hash (for audit trail)
try:
    sha = hashlib.sha256(open(CSV_PATH,'rb').read()).hexdigest()
except FileNotFoundError:
    sha = None
print('Dataset:', CSV_PATH)
print('SHA256 :', sha)
print('Python :', platform.python_version())
print('Pandas :', pd.__version__)
```

## Load data

```python
df = pd.read_csv(CSV_PATH)
assert 'datetime' in df.columns and 'execution_time' in df.columns, 'CSV must contain datetime and execution_time columns'

df['datetime'] = pd.to_datetime(df['datetime'])
# Normalize boolean flags if present
for col in ['is_slow','is_very_slow','is_critical']:
    if col in df.columns and df[col].dtype != bool:
        df[col] = df[col].astype(str).str.lower().map({'true':True,'false':False})

print('Rows   :', len(df))
print('First Row  :', df['datetime'].min())
print('Last Row   :', df['datetime'].max())
print('Agents :', df['agent_type'].nunique(), sorted(df['agent_type'].unique()))
```

## Parameters | Baseline/Peak Windows (inclusive start, exclusive end)

```python
baseline_start = datetime(2025, 11, 6, 21, 30)
baseline_end   = datetime(2025, 11, 6, 22, 00)
peak_start     = datetime(2025, 11, 6, 22, 20)
peak_end       = datetime(2025, 11, 6, 23, 10)

print('Baseline :', baseline_start, '->', baseline_end)
print('Peak     :', peak_start, '->', peak_end)
```

## Standard Deviation Formula:
### σ = √(Σ(x-μ)²/n) (population std)

## Verification | Baseline vs Peak

```python
baseline = df[(df['datetime']>=baseline_start) & (df['datetime']<baseline_end)]
peak     = df[(df['datetime']>=peak_start)   & (df['datetime']<peak_end)]

print(f'Baseline count: {len(baseline):,}')
print(f'Peak count    : {len(peak):,}')

def stats(s: pd.DataFrame):
    return {
        'count': int(len(s)),
        'time_min': str(s['datetime'].min()),
        'time_max': str(s['datetime'].max()),
        'mean': float(s['execution_time'].mean()),
        'median': float(s['execution_time'].median()),
        'p95': float(s['execution_time'].quantile(0.95)),
        'std': float(s['execution_time'].std(ddof=0)),
        'min': float(s['execution_time'].min()),
        'max': float(s['execution_time'].max()),
        'slow_pct_gt20': float((s['execution_time']>20).mean()*100.0),
        'critical_pct_gt60': float((s['execution_time']>60).mean()*100.0),
    }

b = stats(baseline)
p = stats(peak)


print("\nBaseline metrics:")
print(json.dumps(b, indent=2))
print("Peak metrics:")
print(json.dumps(p, indent=2))

slowdown = p['mean']/b['mean'] if b['mean'] else np.nan
increase = (p['mean']-b['mean'])/b['mean']*100 if b['mean'] else np.nan
print(f"Slowdown factor (mean): {slowdown:.2f}x")
print(f"Increase vs baseline : {increase:.2f}%")

# Persist exact results for reports
verification = {
    'baseline_window': {'start': str(baseline_start), 'end': str(baseline_end)},
    'peak_window': {'start': str(peak_start), 'end': str(peak_end)},
    'baseline': b,
    'peak': p,
    'slowdown_factor_mean': round(float(slowdown), 3),
    'increase_percent_mean': round(float(increase), 2),
    'dataset_sha256': sha,
}
open('artifact_verification_case.json','w').write(json.dumps(verification, indent=2))
print('Saved -> artifact_verification_case.json')
```

## Attribution | by agent & top PIDs during peak

```python
by_agent_base = baseline.groupby('agent_type').execution_time.agg(
    count='size', mean='mean', median='median', p95=lambda s: s.quantile(0.95), max='max'
)
by_agent_peak = peak.groupby('agent_type').execution_time.agg(
    count='size', mean='mean', median='median', p95=lambda s: s.quantile(0.95), max='max'
)

print('Baseline: 21:30 - 22:00')
display(by_agent_base.round(3))
print('Peak: 22:20 - 23:10')
display(by_agent_peak.round(3))

peak2 = peak.copy()
peak2['exec_time'] = peak2['execution_time']
pid_contrib = peak2.groupby(['agent_type','pid']).agg(
    count=('exec_time','size'),
    mean=('exec_time','mean'),
    sum_exec=('exec_time','sum'),
    max=('exec_time','max')
).sort_values('sum_exec', ascending=False)

display(pid_contrib.head(10).round(3))

by_agent_base.round(3).to_csv('artifact_baseline_by_agent.csv')
by_agent_peak.round(3).to_csv('artifact_peak_by_agent.csv')
pid_contrib.round(3).to_csv('artifact_peak_top_pid.csv')
print('Saved -> artifact_baseline_by_agent.csv, artifact_peak_by_agent.csv, artifact_peak_top_pid.csv')
```

## Per-minute p95 anomalies with robust z-score

```python
# Aggregate per minute
per_min = df.copy()
per_min['minute_ts'] = per_min['datetime'].dt.floor('min')
agg = per_min.groupby('minute_ts').agg(
    n=('execution_time','size'),
    mean_exec=('execution_time','mean'),
    p95_exec=('execution_time', lambda s: s.quantile(0.95)),
    max_exec=('execution_time','max')
)

# Robust z on p95
med = agg['p95_exec'].median()
mad = (agg['p95_exec']-med).abs().median()
agg['robust_z'] = 0 if mad==0 else 0.6745*(agg['p95_exec']-med)/mad

anoms = agg[agg['robust_z']>3].sort_values('p95_exec', ascending=False)
print(f"Anomalous minutes (robust_z>3): {len(anoms)}")
display(anoms.head(20).round(3))

anoms.round(4).to_csv('artifact_anomalous_minutes.csv')
print('Saved -> artifact_anomalous_minutes.csv')
```

```python
# Build the peak window slice (if not already defined)
peak = df[(df['datetime'] >= peak_start) & (df['datetime'] < peak_end)]

# Compute ECDF
x = np.sort(peak['execution_time'].values)
y = np.arange(1, len(x)+1) / len(x) if len(x) else np.array([])

# Thresholds
THRESH_SLOW = 20.0
THRESH_CRIT = 60.0

# Helper to read % completed under a threshold
def pct_le(thresh):
    if len(x) == 0:
        return 0.0
    return (x <= thresh).mean() * 100.0

pct_20 = pct_le(THRESH_SLOW)
pct_60 = pct_le(THRESH_CRIT)

# Optional p99 marker
p99_peak = peak['execution_time'].quantile(0.99) if len(peak) else None
pct_p99 = pct_le(p99_peak) if p99_peak is not None else None

# Plot ECDF
fig, ax = plt.subplots(figsize=(10, 4))
if len(x):
    ax.plot(x, y, color='#2b8cbe', lw=1.8, label='ECDF (peak window)')
else:
    ax.text(0.5, 0.5, 'No data in peak window', ha='center', va='center', transform=ax.transAxes)

# Threshold lines
ax.axvline(THRESH_SLOW, color='#cc7a00', lw=1.4, ls='--', label='20s (Slow)')
ax.axvline(THRESH_CRIT, color='#b30000', lw=1.4, ls='--', label='60s (Critical)')

# p99 line (optional)
if p99_peak is not None:
    ax.axvline(p99_peak, color='#6a51a3', lw=1.4, ls=':', label=f'p99 ≈ {p99_peak:.2f}s')

# Annotations near the lines
ax.text(THRESH_SLOW, 0.05, f'≤20s: {pct_20:.1f}%', color='#cc7a00', ha='left', va='bottom')
ax.text(THRESH_CRIT, 0.12, f'≤60s: {pct_60:.1f}%', color='#b30000', ha='left', va='bottom')
if p99_peak is not None and pct_p99 is not None:
    ax.text(p99_peak, 0.2, f'≤p99: {pct_p99:.1f}%', color='#6a51a3', ha='left', va='bottom')

ax.set_title('Incident Window ECDF — Fraction of Requests Completed by Time')
ax.set_xlabel('Execution time (s)')
ax.set_ylabel('Cumulative fraction (0–1)')
ax.set_xlim(left=0)
ax.set_ylim(0, 1)
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right', frameon=False)
plt.tight_layout()
plt.show()

# Save for slides
fig.savefig('plot_peak_ecdf.png', dpi=160)
print("Saved -> plot_peak_ecdf.png")
```

```python
# Build slices
baseline = df[(df['datetime'] >= baseline_start) & (df['datetime'] < baseline_end)]
peak     = df[(df['datetime'] >= peak_start)     & (df['datetime'] < peak_end)]

def ecdf(values):
    if len(values) == 0:
        return np.array([]), np.array([])
    x = np.sort(values)
    y = np.arange(1, len(x)+1) / len(x)
    return x, y

x_base, y_base = ecdf(baseline['execution_time'].values)
x_peak, y_peak = ecdf(peak['execution_time'].values)

THRESH_SLOW = 20.0
THRESH_CRIT = 60.0

def pct_le(values, thresh):
    if len(values) == 0:
        return 0.0
    return (values <= thresh).mean() * 100.0

pct_base_20 = pct_le(baseline['execution_time'].values, THRESH_SLOW)
pct_base_60 = pct_le(baseline['execution_time'].values, THRESH_CRIT)
pct_peak_20 = pct_le(peak['execution_time'].values, THRESH_SLOW)
pct_peak_60 = pct_le(peak['execution_time'].values, THRESH_CRIT)

fig, ax = plt.subplots(figsize=(10, 4))
if len(x_base):
    ax.plot(x_base, y_base, color='#1b7837', lw=1.8, label=f'Baseline ECDF (≤20s: {pct_base_20:.1f}%, ≤60s: {pct_base_60:.1f}%)')
if len(x_peak):
    ax.plot(x_peak, y_peak, color='#2b8cbe', lw=1.8, label=f'Peak ECDF (≤20s: {pct_peak_20:.1f}%, ≤60s: {pct_peak_60:.1f}%)')

# Threshold lines
ax.axvline(THRESH_SLOW, color='#cc7a00', lw=1.2, ls='--')
ax.axvline(THRESH_CRIT, color='#b30000', lw=1.2, ls='--')

ax.set_title('ECDF Comparison — Baseline vs Incident (Peak)')
ax.set_xlabel('Execution time (s)')
ax.set_ylabel('Cumulative fraction (0–1)')
ax.set_xlim(left=0)
ax.set_ylim(0, 1)
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right', frameon=False)
plt.tight_layout()
plt.show()

fig.savefig('plot_ecdf_baseline_vs_peak.png', dpi=160)
print("Saved -> plot_ecdf_baseline_vs_peak.png")
```
